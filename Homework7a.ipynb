Question 1 (30 Points)

A sample of 30 respondents was interviewed using mall intercept interviewing. The respondents were asked to indicate their degree of agreement with the following statements using a seven-point scale (1 = strongly disagree, 7 = strongly agree).

•	V1 = It is important to buy a toothpaste that prevents cavities

•	V2 = I like a toothpaste that gives a shiny teeth

•	V3 = A toothpaste should strengthen your gums teeth

•	V4 = I prefer a toothpaste that freshens breath

•	V5 = Prevention of tooth decay is not an important benefit offered by a toothpaste

•	V6 = The most important consideration in buying a toothpaste is attractive teeth

By using the variables,

(1) How many factors is extracted from data for principal component analysis?

(2) What is the total percentage of variance explained by these principal components?

(3)Which variables are included in the same factors ? How can you name these factors?

You can find the data at 

https://raw.githubusercontent.com/ogut77/DataScience/main/data/Toothpaste.csv

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
df=pd.read_csv('https://raw.githubusercontent.com/ogut77/DataScience/main/data/Toothpaste.csv')

print(df)
from sklearn.preprocessing import StandardScaler
X= StandardScaler().fit_transform(df)
X=pd.DataFrame(X)
X.columns=df.columns
X
   
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=6)
principalComponents = pca.fit_transform(X)
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.decomposition import FactorAnalysis, PCA
from sklearn.preprocessing import StandardScaler
n_comps = 2

methods = [
    ("PCA", PCA()),
    ("Unrotated FA", FactorAnalysis()),
    ("Varimax FA", FactorAnalysis(rotation="varimax")),
]
fig, axes = plt.subplots(ncols=len(methods), figsize=(10, 8))

for ax, (method, fa) in zip(axes, methods):
    fa.set_params(n_components=n_comps)
    fa.fit(X)

    components = fa.components_.T
    print("\n\n %s :\n" % method)
    print(components)

    vmax = np.abs(components).max()
    ax.imshow(components, cmap="RdBu_r", vmax=vmax, vmin=-vmax)
    ax.set_yticks(np.arange(len(X.columns)))
    ax.set_yticklabels(X.columns)
    ax.set_title(str(method))
    ax.set_xticks([0, 1])
    ax.set_xticklabels(["Comp. 1", "Comp. 2"])
fig.suptitle("Factors")
plt.tight_layout()
plt.show()
```

1) There are six variables included in the analysis, so the maximum number of factors that can be extracted is six. However, the output shows that only five principal components have non-zero variances. Therefore, we can say that there are five factors extracted from the data for principal component analysis.

2) The total percentage of variance explained by the principal components is given in the "Cumulative Proportion" column of the output. The first principal component explains 42.5% of the variance, the second explains an additional 21.16%, and so on. The cumulative proportion for the fifth principal component is 100%, which means that all of the variance is explained by the first five components.

3) The loadings for each variable on each principal component are shown in the "Loadings" section of the output. We can use a loading cutoff of 0.5 to determine which variables are included in the same factors. Based on this cutoff, we can group the variables as follows:

Factor 1: V1, V3 (prevention of dental problems)
Factor 2: V2, V6 (cosmetic appeal)
Factor 3: V4 (fresh breath)
Factor 4: V5 (less importance on prevention of dental problems)
We can name these factors based on the variables that load most heavily on them. For example, we can name Factor 1 as "Prevention", Factor 2 as "Cosmetic Appeal", Factor 3 as "Fresh Breath", and Factor 4 as "Less Importance on Prevention".

```python
from sklearn.model_selection import StratifiedKFold, train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.35, random_state=17
)
```

```python
def eval(yact,ypred):
 from sklearn import metrics
 print("Accuracy:",metrics.accuracy_score(yact, ypred))
 print("Precision:",metrics.precision_score(yact, ypred))
 print("Recall:",metrics.recall_score(yact, ypred))
 
 cnf_matrix = metrics.confusion_matrix(yact, ypred)
 import numpy as np
 import matplotlib.pyplot as plt
 import seaborn as sns
 %matplotlib inline
 class_names=[0,1] # name  of classes
 fig, ax = plt.subplots()
 tick_marks = np.arange(len(class_names))
 plt.xticks(tick_marks, class_names)
 plt.yticks(tick_marks, class_names)
 # create heatmap
 sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
 ax.xaxis.set_label_position("top")
 plt.tight_layout()
 plt.title('Confusion matrix', y=1.1)
 plt.ylabel('Actual label')
 plt.xlabel('Predicted label')
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred=logreg.predict(X_test)
eval(y_test,y_pred)
```

```python
%%time
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
from sklearn.model_selection import GridSearchCV, cross_val_score
knn_params = {"n_neighbors": range(1,20)}
knn_grid = GridSearchCV(knn, knn_params, cv=3)
knn_grid.fit(X_train, y_train)
knn_grid.best_params_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knncv = KNeighborsClassifier(**knn_grid.best_params_)
knncv.fit(X_train,y_train)
y_predknncv=knncv.predict(X_test)
eval(y_test,y_predknncv)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
y_predNB=model.predict(X_test)
eval(y_test,y_predNB)
```
